```python
category_columns = [
    "source",
    "category",
]

bool_columns = [
    "published_to_website",
]

date_columns = [
    "deleted_at",
    "created_at",
    "updated_at",
    "last_connected_at",
    "last_seen_at",

]
str_columns = [
    # "foursquare_id",
    "name",
    "address",
    "ssid",
    "bssid",
    "author_device_id",
    "client_ip",
    "address_en",
    # "google_place_id",
    "uuid",
    "service_info",
    # "category",
    "foursquare_id",
    "google_place_id",
]

# create the casting types dictionary
casting_types = {}
casting_types.update({col: "category" for col in category_columns})
casting_types.update({col: "bool" for col in bool_columns})
casting_types.update({col: "datetime64[ns]" for col in date_columns})
casting_types.update({col: "str" for col in str_columns})

print(hotspots["published_to_website"].value_counts())
hotspots["published_to_website"].apply(lambda x: True if x == "True" else False ).astype(int)

hotspots = hotspots.astype(casting_types)

print(hotspots.select_dtypes(include='category').nunique())
print()
print(hotspots.select_dtypes(include='category'))
```
